```python
# (b) The difference between Deep copy and Shallow copy
```

```python
# Shallow  copy:

# A shallow copy of an object creates a new object, but does not create copies of nested objects which means the objects
# referred by , an original object. Instead, it copies the references to these objects. Therefore, the changes occured into
# nested objects in the copied objects will reflect in the original object because both the copies share references to the
# same nested objects.
```

```python
# Deep Copy:

# A deep copy of an object creates a new object and recurisively copies all objects
# found in the original object. This means that the new object and it's nested objects are
# entirely independent of the original object.
```

```python
# 5. Create a 3D NumPy array of shape (3,3,3) containing random integers between 1 and 10. Perform the following operations:

#    a.) Extract all even integers from array.
#    b.) Extract all odd integers from array.
```

```python
import numpy as np

# Generate a 3x3x3 array with random integers between 1 and 10 (inclusive)
array1 = np.random.randint(1, 10, size=(3, 3, 3))

# Display the generated array
print("Original Array:\n", array1)

# a) Extract all even integers from the array
even_integers = array1[array1 % 2 == 0]

# b) Extract all odd integers from the array
odd_integers = array1[array1 % 2 != 0]

# Display the extracted even and odd integers
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

```python
# 14. Create a time - series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
#Perform the following tasks using Matplotlib:

# a.) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axis (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# b.) Label the x-axis as 'Date'.
# c.) Set the title of the plot as 'Temperature and Humidity Over Time'.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data for Date, Temperature, and Humidity

dates = pd.date_range(start='2024-03-12', periods=100)
temperature = np.random.randint(10, 30, size=100)
humidity = np.random.randint(30, 70, size=100)

# DataFrame

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Title

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
# 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the
#following tasks using Matplotlib:

# a.) Plot a histogram of the data with 30 bins.
# b.) Overlay a line plot representing the normal distribution's probability density function.
# c.) Label the x-axis as 'Value and the y-axis as 'Frequency/Probability'.
# d.) Set the title of the plot as 'Histogram with PDF Overlay'.
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=1000)

# Creating a histogram
plt.hist(data, bins=30, density=True, alpha=0.5, color='blue', edgecolor='black', label='Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
p = np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)
plt.plot(x, p, color='red', linewidth=2, label='PDF')

# x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Title
plt.title('Histogram with PDF Overlay')

plt.legend()
plt.show()
```

```python
# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants),
# add a legend, label the axis , and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Two random arrays
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Define the quadrant of each values

def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    elif x > 0 and y < 0:
        return 'Q4'
    else:
        return 'Axis'

quadrants = [determine_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrants})

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette='deep')

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

```python
# 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
```

```python
from bokeh.plotting import figure, show
from bokeh.models import Title
import numpy as np

# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create a new plot with title and axis labels
plot = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
plot.line(x, y, legend_label="sin(x)", line_width=2)

# Add grid lines
plot.grid.grid_line_alpha = 0.3

# Show the plot
show(plot)
```

```python
# 19. Using Bokeh, generate a bar chart of randomly generated categorical data, colors bars based on their values, add hover tooltips to display exact values,
# label the axes, and set the title as 'Randomly Categorical Bar Chart'.
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

# Generating a random categorical data

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))
source = ColumnDataSource(data=dict(categories=categories, values=values))

plot = figure(x_range=categories, title="Randomly Categorical Bar Chart", x_axis_label='Categories', y_axis_label='Values')
plot.vbar(x='categories', top='values', width=0.5, source=source, color="firebrick", legend_label="Categories")

hover = HoverTool()
hover.tooltips = [("Value", "@values")]
plot.add_tools(hover)

show(plot)
```

```python
# 20. Using plotly , create a basic line plot of a randomly generated dataset, label the axes ,and set the title as 'Simple Line Plot'.
```

```python
import plotly.graph_objects as go
import numpy as np

x = np.linspace(0, 10, 100)
y = np.random.randn(100)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))
fig.update_layout(title='Simple Line Plot', xaxis_title='X Axis', yaxis_title='Y Axis')

fig.show()
```
